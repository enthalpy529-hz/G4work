```python
#import glob
import numpy as np
file_list = glob.glob('*.npy') # 获取当前目录下所有以.npy为后缀的文件路径
for file_path in file_list:
    # 读取文件内容，此处仅作示例，具体读取方式根据文件内容而定
    np_array = np.load(file_path)
    # 对文件内容进行处理
    print(str(file_path)+" "+str(np.max(np_array))+" "+str(np.min(np_array)))
```

```python
import numpy as np
import matplotlib.pyplot as plt
y = np.load('1000.npy')

# 创建一个 numpy 数组
x = np.arange(np.size(y))

# 绘制折线图
plt.plot(x, y)

# 显示图形
plt.show()
```
